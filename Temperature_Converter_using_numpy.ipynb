```python
import numpy as np
```

```python
celsius=np.array([0,20,37,100,-40,56])
farenheit=(celsius*9/5)+32
cold=celsius[celsius<10]
warm=celsius[(celsius>=10)&(celsius<=30)]
hot=celsius[celsius>30]
```

```python
 for c,f in zip(celsius,farenheit):
  print(f"{c:>4} degrees celsius & {f:>7.1f}farenheit")
print(f"The cold temperatures are {cold} ")
print(f"The warm temperatures are {warm}")
print(f"The hot tempearatures are {hot}")
```
