Numpy code for a student grade tracker that stores student score,calculates averages and finds the highest and lowest grade.

```python
import numpy as np
```

```python
students=np.array(["Amani","Busara","Chema","Duara","Esme"])
scores=np.array([56,82,68,73,97])
average=np.mean(scores)
highest=np.max(scores)
lowest=np.min(scores)
std_dev=np.std(scores)
top_student=students[np.argmax(scores)]
low_student=students[np.argmin(scores)]
passed=scores[scores>=70]
failed=scores[scores<70]
```

```python
print(f"The average score is {average}")
print(f"The highest score is {highest} by {top_student}")
print(f"The lowest score is {lowest} by {low_student} ")
print(f"The number of students who passed is {len(passed)}")
```
