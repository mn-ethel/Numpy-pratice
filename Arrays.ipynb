Numpy Arrays

1 dimensional array

```python
import numpy as np
```

```python
a=np.array([1,2,3,4])
print(a)
a.shape #Outputs four which means they are four columns and it is 1 dimensional
```

2 dimensional array

```python
b=np.array([(1,2,3,4),(5,6,7,8)])
print(b)
b.shape #output(2,4) shows 2 dimensional array
```

Specifying a data type

```python
c=np.array([(1,2,3,4),(5,6,7,8)],dtype=float)
print(c)
```

Initial placeholders in arrays

```python
x=np.zeros((4,5)) #Numpy array of zeros
print(x)
```

```python
y=np.ones((4,3)) #Numpy array of ones
print(y)
```

Array of a particular value

```python
m=np.full((4,5),6)
print(m)
```

Identity Matrix

```python
r=np.eye(6)
print(r)
```

Random Values array

```python
j=np.random.random((2,5))
print(j) # Values are between one and zero
```

Random integer values array within specified range

```python
f=np.random.randint(10,100,(6,7))
print(f)
```

```python
g = np.arange(10,30,5) # In steps of 5
print(g)
```

Converting a list to a numpy array

```python
list1=[2,3,4,5,6]
np_array=np.asarray(list1)
print(np_array)
type(np_array)
```
